# Phase Protraits of Coupled ODE

## Positive Eigen Values means Unstable Source. Origin is the unstable source in this case

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
A = np.array([[3, -1],
             [-1, 3]])
```

```python
# Differential Equation is x_dot = Ax
val, vec = np.linalg.eig(A)
```

```python
D = np.array([[val[0], 0],
             [0, val[1]]])

T = vec
```

```python
def lin_ode(x, t):
    y = A @ x
    x1 = y[0, :]
    x2 = y[1, :]
    return [x1, x2]
```

```python
x1 = np.linspace(-10.0, 10.0, 25)
x2 = np.linspace(-10.0, 10.0, 25)
```

```python
X1, X2 = np.meshgrid(x1, x2)
```

```python
t = 0
u, v = np.zeros(X1.shape), np.zeros(X2.shape)

NI, NJ = X1.shape
```

```python
for i in range(NI):
    for j in range(NJ):
        x = X1[i, j]
        y = X2[i, j]
        q = np.array([[x], [y]])
        yprime = lin_ode( q, t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     

Q = plt.quiver(X1, X2, u, v, color='r')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.xlim([-8, 8])
plt.ylim([-8, 8])
```
