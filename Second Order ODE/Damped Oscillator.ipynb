# Solution of Damped Harmonic Oscillator using Forward Euler

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
```

```python
w = 2*np.pi # Natural Frequency
zeta = 0.3  # Damping ratio (zeta<1 -> Underdamped, zeta>1 -> Overdamped & zeta=1 -> Critically damped)
m = 1 # mass


A = np.array([[0, 1], [-w**2, -2*w*zeta]])

dt = 0.025  # Time Step
T = 10  # Total Time
n = int(T/dt)  # Number of Steps

t = np.linspace(0, T, n)

x0 = [3, 2]  # Initial State, x = 3 & x_dot = 0

# Forward Euler
x_t = np.zeros((2, n))
x_t[:, 0] = x0
for i in range(n-1):
    x_t[:, i+1] = (np.eye(2) + dt * A) @ x_t[:, i]  # x_{i+1} = x_i + h*f(t_i, x_i)
    
plt.figure(figsize=(20, 4))
plt.subplot(1, 3, 1)
plt.plot(t, x_t[0, :], 'k')
plt.xlabel('Time [s]')
plt.ylabel('Position [m]')
plt.legend(['Forward Euler'])
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x_t[0, :], x_t[1, :], 'k')
plt.xlabel('Position [m]')
plt.ylabel('Velocity [m/s]')
plt.legend(['Forward Euler'])
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(t, x_t[1, :], 'k')
plt.xlabel('Time [s]')
plt.ylabel('Velocity [m/s]')
plt.legend(['Forward Euler'])
plt.grid(True)
```
