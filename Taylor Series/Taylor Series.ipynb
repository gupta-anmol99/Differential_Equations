# Plotting first, third, fifth and seventh order Taylor expansion of sin(x)

```python
x = np.linspace(-10, 10, 2000)
y = np.sin(x)
plt.plot(x, y)
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# First Order
P = [1, 0] # This means x + 0
y1 = np.polyval(P, x)
plt.plot(x, y1)

# Third Order
P = [-1/factorial(3), 0, 1, 0]
y3 = np.polyval(P, x) # This means -x^3/3! +x +0
plt.plot(x, y3)

# Fifth Order
P = [1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
y5 = np.polyval(P, x) 
plt.plot(x, y5)

# Seventh Order
P = [-1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
y7 = np.polyval(P, x)
plt.plot(x, y7)

# Ninth Order
P = [1/factorial(9), 0, -1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
y9 = np.polyval(P, x)
plt.plot(x, y9)

# Eleventh Order
P = [-1/factorial(11), 0, 1/factorial(9), 0, -1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
y11 = np.polyval(P, x)
plt.plot(x, y11)

plt.legend( ['sin(x)', '1st Order', '3rd Order', '5th Order', '7th Order', '9th Order', '11th Order'])
```
